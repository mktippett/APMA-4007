# Lecture 1

Take-away message
1. Vectors can be scaled and added to form a linear combination.
$$ x \vec{a}_1  + y \vec{a}_2 = \text{linear combination}$$
2. A system of linear linear equation states that one vector is a linear combination of other vectors.
$$ x \vec{a}_1  + y \vec{a}_2 = \vec{b} $$
3. The scaling factors $x$ and $y$ are the solution.
4. The solution exists if $\vec{b}$ is a linear combination of $\vec{a}_1$ and $\vec{a}_2$

## Example

System of linear equations

$$ 2x + y = 5$$
$$ x - y = 1 $$

Take a moment to solve with a collegue. 

What is the *method*?
What is the *picture*?

## The row picture

Each row is the equation of a line.

The solution $(x, y)$ is on both lines. Plot the lines and find where they cross.

What are the possibilities?

```python
x = np.linspace(-2, 4)
y1 = 5 - 2*x
y2 = x - 1
plt.plot(x, y1, label='y = 5 - 2x')
plt.plot(x, y2, label='y = x - 1')
plt.plot(2, 1, 'o', label='solution')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

## The column picture

Write the system of linear equations as a vector equation:

$$ \begin{bmatrix} 2x + y \\ x - y \end{bmatrix} 
= 
\begin{bmatrix} 5 \\ 1 \end{bmatrix} $$

$$ x \begin{bmatrix} 2 \\ 1 \end{bmatrix} 
 + y \begin{bmatrix} 1 \\ -1 \end{bmatrix}
 = \begin{bmatrix} 5 \\ 1 \end{bmatrix} $$
 
The vector $\begin{bmatrix} 5 \\ 1 \end{bmatrix} $ is what linear combination of two given vectors?

```python
a1 = np.array([2, 1])
a2 = np.array([1, -1])
2 * a1 + 1 * a2
```

```python
def plot_vector(start, v, label=None):
    end = start + v
    plt.plot((start[0], end[0]), (start[1], end[1]), label=label)

x = 2
y = 1
zero = np.zeros(2)

plot_vector(zero, x * a1, label='$x a_1$')
plot_vector(zero, y * a2, label='$y a_2$')
plot_vector(y * a2, x * a1)
plot_vector(x * a1, y * a2)
plot_vector(zero, x * a1 + y * a2, label='$x a_1 + y a_2 $')
plt.legend()
plt.grid()
plt.show()
```
